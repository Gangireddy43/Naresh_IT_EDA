$ Normalization $

- minmaxscalar

```python
# x-x_min/(x_max-x_min)
#step1 : read the pwage columnabs
# step2 : find the min value of the pwage column
# step3 : find the max value of the pwage column
# step4 : nr=datacolumn-min value 
# step5  : dr=max_value-min_value
# step6 : nr/dr


path =r'C:\Users\REDDY\DATA SCIENCE\python datafiles\Visadataset - Visadataset.csv'
visa_df=pd.read_csv(path)
x=visa_df['prevailing_wage']
x_min=visa_df['prevailing_wage'].min()
x_max=visa_df['prevailing_wage'].max()
num=x-x_min
din=x_max-x_min
#min_max_pre=num/din
visa_df['min_max_pre']=num/din
```
